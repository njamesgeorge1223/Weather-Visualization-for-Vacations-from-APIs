# WeatherPy

```python
 #*******************************************************************************************
 #
 #  File Name:  WeatherPy.ipynb 
 #
 #  File Description:
 #      This interactive Python notebook, WeatherPy.ipynb, uses a Python script to 
 #      visualize the weather of over 500 cities of varying distances from the equator. 
 #      The script uses the citipy Python library and the openWeatherMap API to create 
 #      a representative model of weather across cities.
 #      
 #
 #  Date            Description                             Programmer
 #  ----------      ------------------------------------    ------------------
 #  08/12/2023      Initial Development                     N. James George
 #
 #******************************************************************************************/

import PyFunctions as function
import PySubRoutines as subroutine
import PyLogFunctions as log_function
import PyLogSubRoutines as log_subroutine
import WeatherPyAPIFunctions as api_function
import WeatherPyConstants as local_constant
import WeatherPyFunctions as local_function

import pandas as pd
```

```python
CONSTANT_LOCAL_FILE_NAME \
    = 'WeatherPy.py'


log_subroutine \
    .SetLogMode \
        (False)

log_subroutine \
    .SetDebugMode \
        (False)

log_subroutine \
    .SetImageMode \
        (False)


log_subroutine \
    .BeginProgramExecution \
        ('WeatherPy')
```

# <br> **Section 1: Weather Data Acquisition**

## **1.1: List of Cities**

```python
citiesListOfStrings \
    = api_function \
        .ReturnCitiesList()
```

## **1.2: Weather Data from List of Cities**

```python
cityWeatherDataFrame \
    = api_function \
        .ReturnWeatherDataFrame \
            (citiesListOfStrings)

log_function \
    .DebugReturnObjectWriteObject \
        (cityWeatherDataFrame)
```

## **1.3: City Weather Data Set Records Count**

```python
cityWeatherCountIntegerVariable \
    = cityWeatherDataFrame \
        ['City'] \
            .count()

print \
    (f"There are {cityWeatherCountIntegerVariable} " \
     + f"records in the city weather DataFrame.")
```

## **1.4: Display City Weather Data Set**

```python
captionString \
    = 'Table: 1.4: City Weather Information'

currentStylerObject \
    = local_function \
        .ReturnCityWeatherStylerObjectStandardFormat \
            (cityWeatherDataFrame.head(12),
             captionString)

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```

## **1.5: Display City Weather Locations**

```python
hoverColumnsListOfStrings \
    = ['Latitude', 
       'Longitude', 
       'City', 
       'Humidity', 
       'Hotel Name', 
       'Country']

function \
    .DisplayHVPlotFromDataFrame \
        (cityWeatherDataFrame,
         'Figure 1.5: City Weather Locations',
         'City',
         'Humidity',
         (-180, 180),
         (-55, 75),
         0.7,
         'OSM',
         hoverColumnsListOfStrings)
```

## **1.6: Export to CSV File**

```python
subroutine \
    .WriteDataFrameToCSVFile \
        (cityWeatherDataFrame,
         local_constant \
             .WEATHER_DATA_FILE_PATH,
         local_constant \
             .WEATHER_DATA_FILE_INDEX_NAME)
```

## **1.7: Data Import from CSV File**

```python
sampleCityWeatherDataFrame \
    = function \
        .ReturnCSVFileAsDataFrame \
            (local_constant \
                 .WEATHER_DATA_FILE_PATH,
             local_constant \
                 .WEATHER_DATA_FILE_INDEX_NAME)
```

## **1.8: Display Imported City Weather Data Set**

```python
captionString \
    = 'Table: 1.8: Imported City Weather Information'

currentStylerObject \
    = local_function \
        .ReturnCityWeatherStylerObjectStandardFormat \
            (sampleCityWeatherDataFrame.head(12),
             captionString)

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```

## **1.9: Display Imported City Weather Locations**

```python
function \
    .DisplayHVPlotFromDataFrame \
        (cityWeatherDataFrame,
         'Figure 1.9: Imported City Weather Locations',
         'City',
         'Humidity',
         (-180, 180),
         (-55, 75),
         0.7,
         'OSM',
         hoverColumnsListOfStrings)
```

## **Discussion about the linear relationship:** 
### Although this assignment is about linear regression, I notice that the independent variable, latitude, has a relationship with the dependent variable, temperature, just not a linear one.  With linear regression, we obtain a -0.1346 correlation; with quadratic regression, the correlation increases to 0.7879.  In practiice, analysts consider a 0.60-0.79 correlation to be strong.  What does this mean?  The maximum temperature is hottest at the equator and cools down as we move towards either the North Pole or the South Pole.  On top of that, while correlation tells us there is a relationship between two variables, it does not indicate cause and effect.

## **Discussion about the linear relationship:** 
### From the results of the linear regression, in the northern hemisphere, the latitude is negatively correlated with temperature; this negative correlation tells us that the two variables move in opposite directions: as latitude increases, the temperature drops, and vice versa.  In the southern hemisphers, the latitude is positively correlated with temperature: this positive correlation tells us that the two variables move in the same direction: as latitude increases, the temperature increases.  What does this mean?  The maximum temperature is hottest at the equator and cools down as we move towards either the North Pole or the South Pole.  On top of that, while correlation tells us there is a relationship between two variables, it does not indicate cause and effect.
