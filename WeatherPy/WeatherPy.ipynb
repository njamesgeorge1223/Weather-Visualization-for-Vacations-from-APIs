# <br> **Section 1: Weather Variables and Latitude**

## **Section 1.1: Data Acquisition**

### **List of Cities**

```python
citiesListOfStrings \
    = api_function \
        .ReturnCitiesList()
```

### **Weather Data from List of Cities**

```python
cityWeatherDataFrame \
    = api_function \
        .ReturnWeatherDataFrame \
            (citiesListOfStrings)
```

```python
log_function \
    .DebugReturnObjectWriteObject \
        (cityWeatherDataFrame)
```

### **Records Count**

```python
cityWeatherCountIntegerVariable \
    = cityWeatherDataFrame \
        ['City'] \
            .count()

print \
    (f"There are {cityWeatherCountIntegerVariable} " \
     + f"records in the city weather DataFrame.")
```

### **Display City Weather DataFrame**

```python
local_function \
    .ReturnCityWeatherStylerObjectStandardFormat \
        (cityWeatherDataFrame,
         'Table: 1.1: City Weather Information')
```

### **Data Export to CSV File**

```python
subroutine \
    .WriteDataFrameToCSVFile \
        (cityWeatherDataFrame,
         local_constant.WEATHER_DATA_FILE_PATH,
         local_constant.WEATHER_DATA_FILE_INDEX_NAME)
```

```python
sampleCityWeatherDataFrame \
    = function \
        .ReturnCSVFileAsDataFrame \
            (local_constant.WEATHER_DATA_FILE_PATH,
             local_constant.WEATHER_DATA_FILE_INDEX_NAME)
```
